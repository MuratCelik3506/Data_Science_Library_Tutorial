```python
import networkx as nx
import matplotlib.pyplot as plt
import nxviz as nv
import pandas as pd
import numpy as np
```

```python
# read from mtx data
G = nx.read_edgelist("data/dolphin.mtx")
```

```python
nx.draw(G,with_labels=True)
plt.show()
```

```python
dolphin_info = list()
```

```python
len_node = len(G.nodes())
dolphin_info.append(len_node)
len_node
```

```python
len_edge = len(G.edges())
dolphin_info.append(len_edge)
len_edge
```

```python
density = 2*len_edge/(len_node*(len_node-1))
dolphin_info.append(density)
density
```

```python
d_max = sorted(G.degree, key=lambda x: x[1], reverse=True)[0]
dolphin_info.append(d_max[1])
d_max[1]
```

```python
sum_avg = 0
len_avg = 0
for x in G.degree:
    sum_avg += x[1]
    len_avg+=1
d_avg = sum_avg/len_avg
dolphin_info.append(d_avg)
d_avg
```

```python
max_k_core = max(nx.algorithms.core.core_number(G).values())
dolphin_info.append(max_k_core)
max_k_core
```

```python
clustring_coeff = nx.algorithms.cluster.average_clustering(G)
dolphin_info.append(clustring_coeff)
clustring_coeff
```

```python
Diam_G = nx.algorithms.distance_measures.diameter(G)
dolphin_info.append(Diam_G)
Diam_G
```

```python
dist_avg = nx.average_shortest_path_length(G)
dolphin_info.append(dist_avg)
dist_avg
```

```python
tri_num = 0
for k,v in nx.algorithms.cluster.triangles(G).items():
    tri_num += v
tri_num = tri_num/3
dolphin_info.append(tri_num)
tri_num
```

```python
sum_avg = 0
len_avg = 0
for k,v in nx.algorithms.centrality.eigenvector_centrality(G).items():
    sum_avg += v
    len_avg+=1
eigen_vector_centrality = sum_avg/len_avg
dolphin_info.append(eigen_vector_centrality)
eigen_vector_centrality
```

```python
df = pd.DataFrame(dolphin_info,columns=["dolphin"],index=["|V|","|E|","density(p)","d_max","d_avg","Max_K_Core","C(G)","Diam(G)","dist_avg","T(G)","EC_avg"],dtype=float)
```
