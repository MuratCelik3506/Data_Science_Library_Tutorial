## GroupBy: Split, Apply, Combine
Simple aggregations can give you a flavor of your dataset, but often we would prefer
to aggregate conditionally on some label or index: this is implemented in the socalled
groupby operation. The name “group by” comes from a command in the SQL
database language, but it is perhaps more illuminative to think of it in the terms first
coined by Hadley Wickham of Rstats fame: split, apply, combine.

```python
df.groupby("year")
```

```python
df.groupby("year").sum()
```

```python
df.head()
```

```python
df.groupby("state/region")["population"].mean()
```

### Aggregate, filter, transform, apply

```python
df.groupby('year').aggregate(['min', np.median, max])
```

```python
df.groupby('year').aggregate([np.mean,np.std])
```

```python
def filter_func(x):
    return x['population'].std() > 5
df_A = df.copy()
df_A["std"] = df["population"].std()
df_A.groupby('year').filter(filter_func)
```

```python
df.groupby('year').transform(lambda x : x- x.mean())
```

```python
def norm_by_data2(x):
# x is a DataFrame of group values
    x['population'] -= x['population'].min()
    return x
df.groupby('year').apply(norm_by_data2)
```
